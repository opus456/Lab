### 连续求导的学习

```python
import torch 
x = torch.arange(4,dtype=torch.float32,requires_grad=True) # 加上requires_grad就可以随时看到x的梯度
x,x.grad
```

```python
y = 2 * torch.dot(x,x) # 这里是内积使用每一项项城相加 == y = sum(x*x)
y
```

```python
y.backward()
x.grad # 查看y每次求导对应的x的梯度
```

```python
x.grad.zero_() # 梯度清零 
z = x.sum()
z.backward()
x.grad
```

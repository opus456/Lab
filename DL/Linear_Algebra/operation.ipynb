#### 点积 
两个相同大小的向量进行各项的相乘最后结果进行求和 -- **只能对于向量来说**
对于多维的不行了

```python
y = torch.ones(4,dtype=torch.float32)
x = torch.arange(4,dtype=torch.float32)
x,y,torch.dot(x,y)
```

```python
torch.sum(x * y) # 和上面的效果是一样的
```

#### 范数
L2范数是向量元素平方和的平方根-- $||x_2|| = \sqrt{\sum\limits_{i=1}^n{x_i^2}}$

L1范数每个元素的绝对值求和-- $||x||_1 = \sum\limits_{i=1}^n{|x_i|}$

矩阵的F范数是矩阵元素的平方和的平方根-- $||X||_F = \sqrt{\sum\limits_{i=1}^m\sum\limits_{j=1}^n{x_{ij}^2}}$

```python
u = torch.tensor([3.0,-4.0])
m = torch.tensor([[3.0,4.0],[5.0,6.0]])
u.norm() , torch.norm(u) , torch.abs(u).sum(),m.norm(),torch.norm(m)
```

### axis理解

```python
test = torch.ones(24,dtype=torch.float32).reshape(2,3,4)
test, test.sum(axis=0) , test.sum(axis=1) , test.sum(axis=2)
```

```python
# 如果这里是keepdims的话只是把对应的维度变成了1
a = test.sum(axis=0,keepdims=True)
b = test.sum(axis=1,keepdims=True) 
c = test.sum(axis=2,keepdims=True)
a,b,c,a.shape,b.shape,c.shape
```

```python
# 也是可以对多个维度进行求和的
aa = test.sum(axis=[0,1],keepdims=True)
aa
```

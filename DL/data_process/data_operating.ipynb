### 将多个张量连接起来

```python
x = torch.arange(12,dtype=torch.float32).reshape(3,4)
y = torch.tensor([[1,2,3,4],[2,3,4,5],[3,4,5,6]])
torch.cat((x,y),dim=0), torch.cat((x,y),dim=1) # dimensino维度:0的话按照行,1按照列
```

### 内存问题
使用[:]

```python
# 错误操作
Y = torch.arange(12,dtype=torch.float32).reshape(3,4)
X = torch.arange(12,dtype=torch.float32).reshape(3,4)
before = id(Y) # Y的地址
Y = Y+X
before,id(Y) ,before==id(Y) #发现两个不一样
```

```python
# 正确操作
Z = torch.zeros_like(Y)
print("before id: ",id(Z))
Z[:] = X+Y #必须要使用[:]才能够正确的操作
print("before id: ",id(Z))
```
